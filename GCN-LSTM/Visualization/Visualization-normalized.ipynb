```python
def load_all_pkl_files(folder_path):
    # Dictionary to store the loaded data
    all_models_results = {}

    # List all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is a .pkl file
        if filename.endswith('.pkl'):
            # Full path to the file
            file_path = os.path.join(folder_path, filename)
            
            # Load the .pkl file
            with open(file_path, 'rb') as file:
                model_results = pickle.load(file)
            
            # Use the filename (without extension) as the key in the dictionary
            model_name = os.path.splitext(filename)[0]
            all_models_results[model_name] = model_results
    
    return all_models_results
```

```python
all_models_results = load_all_pkl_files(results_folder)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Function to calculate mean and std for each game
def calculate_mean_std(all_models_results):
    game_results = {1: {}, 2: {}, 3: {}}
    
    for model_name, results in all_models_results.items():
        game = results['game']
        residue_all_samples = results['residue_all_samples']
        mean_residue = np.mean(residue_all_samples)
        std_residue = np.std(residue_all_samples)
        
        model_key = f"{results['model']}_{results['aggregation']}"
        game_results[game][model_key] = {
            'mean': mean_residue,
            'std': std_residue
        }
    
    return game_results

# Calculate mean and std for each game
game_results = calculate_mean_std(all_models_results)

# Ensure all games have the same models
common_models = set(game_results[1].keys()).intersection(game_results[2].keys()).intersection(game_results[3].keys())

# Create a DataFrame with the results and standard deviations
df = pd.DataFrame({
    'Model': list(common_models),
    'Game 1 Mean': [game_results[1][model_name]['mean'] for model_name in common_models],
    'Game 1 Std': [game_results[1][model_name]['std'] for model_name in common_models],
    'Game 2 Mean': [game_results[2][model_name]['mean'] for model_name in common_models],
    'Game 2 Std': [game_results[2][model_name]['std'] for model_name in common_models],
    'Game 3 Mean': [game_results[3][model_name]['mean'] for model_name in common_models],
    'Game 3 Std': [game_results[3][model_name]['std'] for model_name in common_models]
})

df.set_index('Model', inplace=True)
df = df.sort_index()

# Plot the multibar horizontal bar chart with error bars
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.25
index = np.arange(len(df))

# Plot Game 1
ax.barh(index, df['Game 1 Mean'], bar_width, xerr=df['Game 1 Std'], label='Game 1', color='skyblue', capsize=5)
# Plot Game 2
ax.barh(index + bar_width, df['Game 2 Mean'], bar_width, xerr=df['Game 2 Std'], label='Game 2', color='lightgreen', capsize=5)
# Plot Game 3
ax.barh(index + 2 * bar_width, df['Game 3 Mean'], bar_width, xerr=df['Game 3 Std'], label='Game 3', color='salmon', capsize=5)

# Add labels and title
ax.set_xlabel('Mean Residue')
ax.set_title('Mean Residue for Different Models with Error Bars')
ax.set_yticks(index + bar_width)
ax.set_yticklabels(df.index)
ax.legend()

plt.show()

df
```

```python
# Function to normalize predictions
def min_max_normalize(array):
    min_val = array.min()
    max_val = array.max()
    return (array - min_val) / (max_val - min_val)

# Normalize predictions for each model in all_models_results
for model in all_models_results.keys():
    predictions = all_models_results[model]['predicted_influences_all_samples']
    normalized_predictions = min_max_normalize(predictions)
    all_models_results[model]['normalized_predicted_influences'] = normalized_predictions
    print(f"Model: {model}")
    print(f"Original min: {predictions.min()}, Original max: {predictions.max()}")
    print(f"Normalized min: {normalized_predictions.min()}, Normalized max: {normalized_predictions.max()}")
```

```python
# Function to calculate residue
def calculate_residue(true, pred, epsilon=1e-10):
    residue = np.abs(true - pred) / (true + epsilon)
    residue[true == 0] = 0
    return residue

# Function to calculate mean residue per player
def mean_residue_per_player(residue):
    return np.mean(residue, axis=(0, 1))

# Function to calculate average residue all players
def average_residue_all_players(mean_residue):
    return np.mean(mean_residue)

# Function to calculate median residue per player
def median_residue_per_player(residue):
    return np.median(residue, axis=(0, 1))

# Process each model
for model in all_models_results.keys():
    true_influences = all_models_results[model]['true_influences_all_samples']
    normalized_pred_influences = all_models_results[model]['normalized_predicted_influences']
    
    # Calculate residues
    residue_all_samples = calculate_residue(true_influences, normalized_pred_influences)
    mean_residue = mean_residue_per_player(residue_all_samples)
    avg_residue_all_players = average_residue_all_players(mean_residue)
    median_residue = median_residue_per_player(residue_all_samples)
    
    # Store the calculated values
    all_models_results[model]['residue_all_samples_normalized'] = residue_all_samples
    all_models_results[model]['mean_residue_per_player_normalized'] = mean_residue
    all_models_results[model]['average_residue_all_players_normalized'] = avg_residue_all_players
    all_models_results[model]['median_residue_per_player_normalized'] = median_residue

example_model = 'base_add_2'
print("Residue All Samples Normalized:", all_models_results[example_model]['residue_all_samples_normalized'])
print("Mean Residue Per Player Normalized:", all_models_results[example_model]['mean_residue_per_player_normalized'])
print("Average Residue All Players Normalized:", all_models_results[example_model]['average_residue_all_players_normalized'])
print("Median Residue Per Player Normalized:", all_models_results[example_model]['median_residue_per_player_normalized'])
```
